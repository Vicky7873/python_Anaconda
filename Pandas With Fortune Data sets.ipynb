```python
players['batsman_runs'].sum().sort_values(ascending=False).head(10)
# here we got the top batsman from the list
# 1. we make the group of players
# 2. then we just fetc the batsman_run column and s um their runs
# 3. then we sort them and get the head of the columns
```

```python
# who win the most in ipl
data=pd.read_csv('matches.csv')
data['winner'].value_counts()
```

```python
death_over=delivery[delivery['over']>15]
```

```python
players=death_over.groupby('batsman')['batsman_runs'].count()
x=players>200
players[x]
```

```python
# merge 
# multiple columns and the datas are different and we need to merge and made the required column
```

```python
new=delivery.merge(data,left_on='match_id',right_on='id')
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batsman']]
# we have to create multiple 
```

```python
mask=delivery['batsman_runs']==6
six=delivery[mask]
six.shape
```

```python
six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
# this basically shows us how many sixes are beaten by the team in the particular over
```
