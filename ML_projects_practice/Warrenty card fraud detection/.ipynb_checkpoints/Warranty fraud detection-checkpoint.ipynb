```python
Q1 = df['Claim_Value'].quantile(0.25)
Q3 = df['Claim_Value'].quantile(0.75)

IQR = Q3 - Q1

df = df[~((df['Claim_Value'] < (Q1 - 1.5 * IQR)) |(df['Claim_Value'] > (Q3 + 1.5 * IQR)))]
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
sns.boxplot(x = 'Fraud', y = 'Claim_Value', data = df, ax =ax[0]).set_title('Claim Value and Fraudulent Claims')

sns.violinplot(x = 'Fraud', y = 'Claim_Value', data = df, ax =ax[1]).set_title('Claim Value and Fraudulent Claims')
```

# Train Test split

```python
X=df.drop(columns='Fraud')
y=df['Fraud']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=1)
```

```python
df.head(2)
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(C= 0.001, penalty='l1', solver= 'saga')
```

```python
lr.fit(X_train,y_train)

#training accuracy
print(lr.score(X_train,y_train))
```

```python
lr_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(lr_pred,y_test)
```

```python
sns.heatmap(confusion_matrix(lr_pred,y_test),annot=True)
```

```python
lr_param={
    'penalty':['l1', 'l2', 'elasticnet'],
    'C':[0.001,0.002,0.003],
    'solver':['lbfgs', 'iblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}
```

```python
from sklearn.model_selection import GridSearchCV
lr_grd=GridSearchCV(lr,param_grid=lr_param,cv=5)
```

```python
lr_grd.fit(X_train,y_train)
```

```python
lr_grd.best_params_
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(criterion='gini',max_depth= 2,
 min_samples_leaf= 2,
 min_samples_split=2,
 random_state= 0)
```

```python
rfc_param={
    'max_depth': [2,4,6,8],
    'min_samples_leaf': [2,4,6,8],
    'min_samples_split': [2,4,6,8],
    'criterion': ['gini', 'entropy'],
    'random_state': [0,42]
}
```

```python
rfc_grd=GridSearchCV(rfc,param_grid=rfc_param,cv=5)
```

```python
rfc_grd.fit(X_train,y_train)
```

```python
rfc_grd.best_params_
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc.score(X_train,y_train)
```

```python
accuracy_score(rfc.predict(X_test),y_test)
rfc_pred=rfc.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(rfc_pred,y_test),annot=True)
```
