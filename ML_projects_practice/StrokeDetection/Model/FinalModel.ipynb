```python
num_cols=['avg_glucose_level','bmi']
cat_cols=['ever_married','work_type','Residence_type','smoking_status','gender']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df[cat_cols]=df[cat_cols].apply(le.fit_transform)
df.head()
```

```python
df['work_type'].nunique()
```

```python
from sklearn.preprocessing import StandardScaler
stc=StandardScaler()
```

```python
df[num_cols]=stc.fit_transform(df[num_cols])
```

```python
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.shape
```

```python
X=df.drop(columns=['stroke'],axis=1)
X.shape
```

```python
df['stroke'].dtype
```

```python
df['stroke']=df['stroke'].astype('int')
```

```python
df['stroke'].dtype
```

```python
y=df['stroke']
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=30)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc.score(X_train,y_train)
```

```python
rfc_predict=rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_acc=accuracy_score(rfc_predict,y_test)
```

```python
test_acc
```
