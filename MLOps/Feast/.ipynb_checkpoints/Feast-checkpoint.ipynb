# feast is to store the features for ML models

```python
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head(2)
```

```python
predictors_df = df.loc[:,df.columns != 'Outcome']
target_df = df['Outcome']
```

```python
predictors_df.head(2)
```

### Why we need time stamp for predictiong the current data, and we required the time for prediction when the data get store

```python
timestamps = pd.date_range(end = pd.Timestamp.now(),
                           periods = len(df),freq = 'D').to_frame(name = 'event_timestamp', index = False)
```

```python
timestamps
```

```python
# Concate the data with time stamp
```

```python
predictors_df = pd.concat(objs=[predictors_df,timestamps],axis=1)
target_df = pd.concat(objs=[target_df,timestamps],axis=1)
```

```python
predictors_df.head(2)
```

```python
target_df.head(2)
```

# Create an Unique ID for the data set for prediction

```python
df_len = len(df)
len_list = list(range(df_len))
```

```python
len_list
```

```python
patient_ids = pd.DataFrame(data=len_list,columns=['patient_id'])
```

```python
predictors_df = pd.concat(objs=[predictors_df,patient_ids],axis=1)
target_df = pd.concat(objs=[target_df,patient_ids],axis=1)
```

```python
# predictors_df.drop(columns=['patient_id'],axis=0,inplace=True)
```

```python
predictors_df.head(2)
```

```python
target_df.head(2)
```

```python
# target_df.drop(columns='patient_id',axis=0,inplace=True)
```

### Remember Feast accept these fetures of predictors_df in parquet format

```python
predictors_df.to_parquet(path='predictors_df.parquet')
target_df.to_parquet(path='target_df.parquet')
```

### 2 records are create in our folder name as predictors_df.parquet and target_df.parquet
