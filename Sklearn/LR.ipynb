```python
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('ad.csv')
```

```python
data.head()
```

```python
df=data.drop(['Unnamed: 0','Unnamed: 5'],axis=1)
df
```

```python
df2=df.dropna(subset=['TV'])
df2
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(df2['TV'],df2['sales'],c='black')
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
X=df2['TV'].values.reshape(-1,1)
y=df2['sales'].values.reshape(-1,1)
```

```python
reg=LinearRegression()
reg.fit(X,y)
```

```python
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions=reg.predict(X)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(df2['TV'],df2['sales'],c='black')
plt.plot(df2['TV'],predictions,c='blue',linewidth=2)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```
