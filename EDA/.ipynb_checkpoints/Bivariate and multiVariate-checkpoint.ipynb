# 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(tips['total_bill'],tips['tip'])
# here we can see the total bill is high then the tip is also high
```

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'])
# multivariate analysis
# here we plot the genders of all customers
```

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
# get the whihc customer is smoker or not
```

# 2. Bar Plot (Numerical - Categorical)

```python
titanic.columns
```

```python
# find out the avarage age travel in the class
# in x axis we store the category and in y axis we store the numerical columns
sns.barplot(titanic['Pclass'],titanic['Age'])
```

```python
sns.barplot(titanic['Pclass'],titanic['Fare'],hue=titanic['Sex'])
# here the females are buied most expense tickets in the classes
```

```python
# find out how many male and females are high in ages in the classes
sns.barplot(titanic['Pclass'],titanic['Age'],hue=titanic['Sex'])
```

# 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(titanic['Sex'],titanic['Age'],hue=titanic['Survived'])
# here it shows the younger male and older female survived mostly
```


# 4. Distplot (Numerical - Categorical)

```python
# find out the difference of survival in different age group
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
# here we plot the died range on the ages category
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
# here we plot the survived range on the ages category
# here we can analyize that the survival on the age range 0 to 15 is high
# the death range in between 20 to 40 is low
# if your age is more than 60 the survival chance is low
```

# 5. HeatMap (Categorical - Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
# get the value that how many peoples are survived and and died in particular class
```
